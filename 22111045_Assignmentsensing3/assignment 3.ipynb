# Question1: Spectrogram of Audio File

```python
import os
import matplotlib.pyplot as plt
import librosa
import librosa.display
import IPython.display as ipd
import pandas as pd
```

# Reading audio file using librosa

```python
audio_file = "audio_sample.wav"
ipd.Audio(audio_file)
```

```python
y, sr = librosa.load(audio_file)
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveshow(y, sr=sr)
```

```python
X = librosa.stft(y)
S_db = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(S_db, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
plt.figure(figsize=(14, 5))
librosa.display.specshow(S_db, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()
```

## from the spectrogram it can be clearly observed that the spectrogram comprises of five spectrum each corresponding to a word of a audio file

```python
import csv

acceleration_x = []
acceleration_y=[]
acceleration_z=[]
s=[acceleration_x,acceleration_y,acceleration_z]
with open("Testing.txt", 'r') as txtfile:
    lines = csv.reader(txtfile, delimiter=',')

    for rows in lines:
        for i in range(3):
            s[i].append(float(rows[i]))
       
    
for i in range(len(acceleration_x)):
    if ((acceleration_x[i] >= min_standing_x and acceleration_x[i] <= max_standing_x) and (acceleration_y[i] >= min_standing_y and acceleration_y[i] <= max_standing_y) and (acceleration_z[i] >= min_standing_z and acceleration_z[i] <= max_standing_z)):
        print("acceleration_x: ",acceleration_x[i]," acceleration_y: ",acceleration_y[i]," acceleration_z: ",acceleration_z[i],"Standing")
    elif ((acceleration_x[i] >= min_sitting_x and acceleration_x[i] <= max_sitting_x) and (acceleration_y[i] >= min_sitting_y and acceleration_y[i] <= max_sitting_y) and (acceleration_z[i] >= min_sitting_z and acceleration_z[i] <= max_sitting_z)):
        print("acceleration_x: ",acceleration_x[i]," acceleration_y: ",acceleration_y[i]," acceleration_z: ",acceleration_z[i],"Sitting")
    elif ((acceleration_x[i] >= min_lying_x and acceleration_x[i] <= max_lying_x) and (acceleration_y[i] >= min_lying_y and acceleration_y[i] <= max_lying_y) and (acceleration_z[i] >= min_lying_z and acceleration_z[i] <= max_lying_z)):
        print("acceleration_x: ",acceleration_x[i]," acceleration_y: ",acceleration_y[i]," acceleration_z: ",acceleration_z[i],"Sleeping")
    else:
        print("acceleration_x: ",acceleration_x[i]," acceleration_y ",acceleration_y[i]," acceleration_z: ",acceleration_z[i],"Pattern Not Recognized")
```
